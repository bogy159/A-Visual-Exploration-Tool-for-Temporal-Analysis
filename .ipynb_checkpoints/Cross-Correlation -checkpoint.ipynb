```python
from scipy import signal, stats
import matplotlib.pyplot as plt
import numpy as np

sig = np.repeat([0., 1., 1., 0., 1., 0., 0., 1.], 128)
sig2 = np.repeat([0., 0., 1., 1., 0., 1., 0., 0.], 128)
#win = signal.gaussian(64, 50)
#sig_gauss = signal.fftconvolve(sig, win)
sigma = 0.5
sig_noise = sig + (sigma * np.random.randn(len(sig)))
sig2_noise = sig2 + (sigma * np.random.randn(len(sig)))

corr = signal.correlate(sig_noise, sig2_noise, mode='same') 

shift = int(np.argmax(corr) - len(sig) / 2)
shift
```

```python
sig_shift = sig2_noise
sig_shift = np.roll(sig_shift, shift)
sig_shift
```

```python
%matplotlib notebook

fig, (ax_orig, ax_sig2, ax_corr, ax_sig_shift) = plt.subplots(4, 1, sharex=True)

ax_orig.plot(sig_noise)
ax_orig.set_title('Signal 1')

ax_sig2.plot(sig2_noise)
ax_sig2.set_title('Signal 2')

ax_corr.plot(corr)
ax_corr.axhline(0.5, ls=':')
ax_corr.set_title('Cross-correlation of signal 1 and 2')

ax_sig_shift.plot(sig_shift)
ax_sig_shift.set_title('Shifted signal 2')

ax_orig.margins(0, 0.1)

fig.tight_layout()
fig.show()
```

```python
rho = stats.pearsonr(sig_noise, sig_shift)
rho
```
