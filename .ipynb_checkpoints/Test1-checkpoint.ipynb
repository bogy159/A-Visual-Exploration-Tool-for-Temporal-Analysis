```python
import pandas as pd 
import numpy as np
import time
import nltk

import spacy
nlp = spacy.load("en_core_web_sm", disable=['parser', 'ner'])
stopwords = set(nltk.corpus.stopwords.words('english'))
tokenizer = nltk.tokenize.RegexpTokenizer(r'\w+')
```

```python
def readCSV(item):
    data = pd.read_csv(item, dtype={0: int}, index_col=0, encoding='latin-1')
    df = pd.DataFrame(data)
    return df    
```

```python
samRev = readCSV("revNamed3.csv")
print(samRev.shape)
```

```python
samRev1 = samRev.copy()
samRev1
```

```python
def Smapling(item):
    result = item
    result['date'] = pd.to_datetime(result.date)
    result.drop(["neighborhood", "address", "state", "city", "latitude", "longitude", "stars_y", "review_count", "categories", "language"], axis=1, inplace=True)
    result.rename(columns={'stars_x': 'stars'}, inplace=True)
    result['ccur'] = result.groupby('business_id')['business_id'].transform('size')
    result = result[result.ccur > 10]
    result = result.sort_values('ccur', ascending = False)
    
    return result
```

```python
samRev1 = samRev.copy()
samRev1 = Smapling(samRev1)
samRev1
```

```python
pd.DataFrame(samRev1.business_id.value_counts()).reset_index()
```

```python
def tokenise(text):
    dirty = tokenizer.tokenize(text)
    result = [w.lower() for w in dirty if w.lower() not in stopwords]
    return result
```

```python
def lemmatizingSpacy(item):
    
    document = spacy.tokens.Doc(nlp.vocab, words=item)
    result = [w.lemma_ for w in document]
    return result
```

```python
proba1 = samRev1.copy()

start = time.time()

proba1["tokens"] = proba1["text"].apply(tokenise)
t1 = time.time()
print("Token done: " + str(t1 - start))

proba1["lemmas"] = proba1["tokens"].apply(lemmatizingSpacy)
l1 = time.time()
print("Lemma done: " + str(l1 - t1))
proba1
```

```python
proba2 = proba1.copy()
del proba2['tokens']
proba2
```

```python
proba2.to_csv('revSamLem2.csv')
```
