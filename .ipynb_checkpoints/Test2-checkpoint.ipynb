```python
import pandas as pd 
import numpy as np
```

```python
def reviewData(item):
    data = pd.read_csv(item)
    df = pd.DataFrame(data) 
    df.rename(columns=df.iloc[0]).drop(df.index[0])
    df.drop(["review_id", "user_id", "useful", "funny", "cool"], axis=1, inplace=True)
    df = df.sort_values('business_id', ascending = True)
    df['date'] = pd.to_datetime(df.date)
    return df

reviews = reviewData('yelp_review.csv')
reviews
```

```python
def businessData(item):
    data = pd.read_csv(item)    
    df = pd.DataFrame(data) 
    df.rename(columns=df.iloc[0]).drop(df.index[0])
    df.drop(["is_open"], axis=1, inplace=True)
    categories =['Restaurant', 'Food']
    pattern = '|'.join(categories)
    df = df[df['categories'].str.contains(pattern, na = False)]
    return df

businesses = businessData('yelp_business.csv')
businesses
```

```python
def addNames(df1,df2):
    result = df1.merge(df2, on='business_id', how='left')
    return result

revNamed = addNames(reviews,businesses)
revNamed
```

```python
revNamed = revNamed[revNamed.name.notnull()]
revNamed
```

```python
revNamed = revNamed.sort_values('date', ascending = True)
revNamed.to_csv('revNamed.csv')
revNamed
```

```python
def Smapling(item):
    #result = item.sample(n=1000, random_state=1)
    
    #item = item.sort_values('review_count', ascending = False)
    mid = item[item.review_count > 2500]
    print(mid.shape)
    result = mid.sample(n=1000, random_state=1)
    #result = item.head(10000)
    
    #result = result.sort_values('date', ascending = True)
    result['date'] = pd.to_datetime(result.date)
    result.drop(["neighborhood", "address", "city", "state", "postal_code", "latitude", "longitude", "stars_y", "review_count", "categories"], axis=1, inplace=True)
    result.rename(columns={'stars_x': 'stars'}, inplace=True)
    result['ccur'] = result.groupby('business_id')['business_id'].transform('size')
    result = result.sort_values('ccur', ascending = False)
    return result
revSam = Smapling(revNamed)
revSam
```

```python
def oneRest(data,word):
    result = data[data['business_id'].str.contains(word, na = False)]
    #result['date'] = pd.to_datetime(result.date)
    return result

singleRest = oneRest(revSam, "4JNXUYY8wbaaDmk3BPzlWw")
singleRest
#singleRest
```
