```python
import pandas as pd 
import numpy as np
import time
import nltk
import re
from nltk.util import ngrams

import Filter as fil
import SentimentAnalysis as sap


import spacy
nlp = spacy.load("en_core_web_sm", disable=['parser', 'ner'])
stopwords = set(nltk.corpus.stopwords.words('english'))
from textblob import TextBlob

tokenizer = nltk.tokenize.RegexpTokenizer(r'\w+')
```

```python
def readCSV(item):
    data = pd.read_csv(item, dtype={0: int}, index_col=0, encoding='latin-1')
    df = pd.DataFrame(data)
    return df
```

```python
allData = fil.main("", "", "", "", "", "", "")
allData
```

```python
def lemmatizingSpacy(item):
    
    document = spacy.tokens.Doc(nlp.vocab, words=item)
    result = [w.lemma_ for w in document]
    return result
```

```python
def phrasesss(text):
    nlp = spacy.load('en_core_web_sm')
    doc = nlp(text)
    result = [' '.join(lemmatizingSpacy(tokenise(str(chunk).lower()))) for chunk in doc.noun_chunks if (str(chunk).lower() not in stopwords)]
    
    #for chunk in doc.noun_chunks:
       # print(len(chunk))
    return result
```

```python
def tokenise(text):
    dirty = tokenizer.tokenize(text)
    result = [w.lower() for w in dirty if w.lower() not in stopwords]
    return result
```

```python
proba1 = allData.copy()

start = time.time()

proba1["tokens"] = proba1["text"].apply(tokenise)
t1 = time.time()
print("Token done: " + str(t1 - start))

proba1["lemmas"] = proba1["tokens"].apply(lemmatizingSpacy)
l1 = time.time()
print("Lemma done: " + str(l1 - t1))

proba1["phrases"] = proba1["text"].apply(phrasesss)
f1 = time.time()
print("Phrase done: " + str(f1 - l1))

proba1["polarity"] = proba1["text"].apply(sap.detect_polarity)
p1 = time.time()
print("Polar done: " + str(p1 - f1))

end = time.time()
print("All done: " + str(end - start))

#len(proba1.iloc[0]["phrases"])
#proba1.iloc[0]["phrases"]
proba1
```
